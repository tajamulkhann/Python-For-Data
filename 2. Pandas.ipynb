### **2. Series**
`single list with index`  `Combination of Np.Array and Index`

*A Series is very similar to a NumPy array (in fact it is built on top of the NumPy array object). What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be `indexed by a label`, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.*

````python
import numpy as np
import pandas as pd
````

#### **Creating Series**


---




````python
labels = ['a','b','c']
my_list = [10,20,30]
my_tuple = (10,20,30)
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
````

 you can use the .groupby() method to group rows together based off of a column name

`Syntax - df.grouby('col_to_be_grouped').agg_fun`

````python
P = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
col = [                #Adding Column names in
        "age",
        "workclass",
        "fnlwgt",
        "education",
        "educationNum",
        "maritalStatus",
        "occupation",
        "relationship",
        "race",
        "sex",
        "capitalGain",
        "capitalLoss",
        "hoursPerWeek",
        "nativeCountry",
        "Above50K"
]
P.columns = col
P.head()
````

* **Advanced Level Grouping**


---

Syntax
`x.groupby('col_to_be_grouped').agg(new_col_name = ("colname", 'func'), new_col_name2 = ("colname2", 'func2'))`

````python
P.groupby(["education"]).agg(sum_capitalGain = ("capitalGain", 'sum'),
                              sum_capitalLoss= ("capitalLoss", 'sum'),
                              avg_age = ("age", 'mean')).reset_index()
````

### **8. Merging, Joining and Concatenating**

There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating. In this lecture we will discuss these 3 methods with examples.

____

````python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
````

````python
df1
````

````python
df2
````

````python
df3
````

#### 1. **Concatenation**


---



Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on. You can use **pd.concat** and pass in a list of DataFrames to concatenate together:

````python
pd.concat([df1,df2,df3])
````

````python
pd.concat([df1,df2,df3],axis=1) #Along the Column
````

#### 2. **Merge**


---


The **merge** function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example:

`pd.merge(x,y, how= [inner,outer, left, right] , on = [Common Column]`

````python
import numpy as np
import pandas as pd
````

````python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
left
````

````python
right
````

````python
pd.merge(left, right, how= 'inner', on = 'key' )
````

````python
pd.merge(left, right, how= 'inner', on = 'C' )
````

#### 3. **Joining**
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.**bold text**

````python
left1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
````

````python
left1.join(right2)
````

````python
left1.join(right2, how='outer')
````

### **9. Cross Tab and Pivot Tables**


#### **9.1  CrossTab**

---

Aggregation at both Column and Row level unlike Groupby which takes only one argument

`Syntax = pd.crosstab(index = x.col1, columns=x.col2, values=x.age, aggfunc="mean")`

````python
dk = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
col = [                #Adding Column names in
        "age",
        "workclass",
        "fnlwgt",
        "education",
        "educationNum",
        "maritalStatus",
        "occupation",
        "relationship",
        "race",
        "sex",
        "capitalGain",
        "capitalLoss",
        "hoursPerWeek",
        "nativeCountry",
        "Above50K"
]
dk.columns = col
dk.sample(3)
````

````python
pd.crosstab(index = dk.education, columns=dk.sex, values=dk.age, aggfunc="mean")
````

````python
pd.crosstab(index=dk.education, columns=dk.sex, values=dk.age, aggfunc=["mean", 'sum'])
````

#### **9.2  Pivot Tables**

`syntax = pd.pivot_table(data = x, index=x.education, values= ["age"], columns=x.sex, aggfunc=['mean'])`

````python
pd.pivot_table(data = dk, index=dk.education, values= ["age"], columns=dk.sex, aggfunc=['mean'])
````

````python
pd.pivot_table(data = dk, values= ["age", 'capitalGain', 'capitalLoss'], index=dk.education, columns=dk.sex,
               aggfunc={"age":"mean", 'capitalGain': "max", 'capitalLoss': 'min'})
````

````python
pd.pivot_table(data = dk, values=["age", 'capitalGain', 'capitalLoss'], index=dk.education,
               columns = dk.sex, aggfunc={"age": "mean", 'capitalGain': ["mean", "sum", "max"], 'capitalLoss': 'sum'})
````
